```python
import pandas as pd
countries = pd.read_csv("world-happiness-report-2021.csv")
houses = pd.read_csv("kc_house_data.csv")
```

```python
# はまりどころ。inplace=Trueにしないとダメ
countries.set_index("Country name", inplace=True)
```

```python
countries.head()
```

# loc

```python
# Series
countries.loc["Switzerland"]
```

```python
# DataFrame
countries.loc[["Switzerland"]]
```

```python
# indexでsortして、slice
countries.sort_index().loc["Denmark":"France"]
```

```python
countries.loc["Denmark", ["Social support", "Healthy life expectancy"]]
```
