```python
import pandas as pd
```

# from list

```python
ice_cream = ["Chocolate", "Vanilla", "Strawberry", "Rum Raisin"]
pd.Series(ice_cream)
```

# from dict

```python
sushi = {
    "Samon": "Orange",
    "Tuna": "Red",
    "Eel": "Brown"
}
pd.Series(sushi)
```

# from data + index

```python
numbers = [30, 11, 23]
names = ["Steph", "Clay", "Draymond"]
pd.Series(data=names, index=numbers)
```

# series methods

```python
prices = pd.Series([2.99, 4.45, 1.36])
```

```python
prices.sum()
```

```python
prices.mean()
```
