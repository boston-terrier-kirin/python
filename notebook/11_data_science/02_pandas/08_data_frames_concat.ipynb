```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
{
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"],
    "C": ["C0", "C1", "C2"]
},
index=[0, 1, 2])
```

```python
df2 = pd.DataFrame(
{
    "A": ["A3", "A4", "A5"],
    "B": ["B3", "B4", "B5"],
    "C": ["C3", "C4", "C5"]
},
index=[3, 4, 5])
```

```python
df3 = pd.DataFrame(
{
    "A": ["A6", "A7", "A8"],
    "B": ["B6", "B7", "B8"],
    "C": ["C6", "C7", "C8"]
},
index=[6, 7, 8])
```

# concat

## axisは、0:row, 1:col

```python
pd.concat([df1, df2, df3], axis=0)
```

```python
pd.concat([df1, df2, df3], axis=1)
```
